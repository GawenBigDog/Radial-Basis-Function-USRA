##Smoothing with Averages
Smoothing the RBF data sites by interpolating on less centers. New centers are computed by taking the average of $n$ data sites.

```python

#The SciPy interpolate library has a Radial Basis Function called Rbf
from scipy.interpolate import Rbf
from scipy.spatial import KDTree

#This allows us to display iPython Notebook plotting results inside the notebook
%pylab inline





#Function to quickly generate a sample piecewise function
def quickpiece(x):
    return np.piecewise(x,[np.abs(x)<1,np.abs(x)>=1],[1,0])
#np.piecewise(x,[x<1,x>=1],[0,1])

#Function to break list into smaller n-sized chunks
def chunks(l, n):
    for i in xrange(0, len(l), n):
        yield l[i:i+n]
        
#Average the data points from the cluster
def avg(arraylist):
    result=[]
    for array in arraylist:
        result.append(np.average(array))
    return result


#I can define my own RBF function here
#It needs to be a function of self and r
#The epsilon argument is passed by self.epsilon
def my_function(self,r):
    return sqrt((r/self.epsilon)**2 + 1)
    
    
#Defining the data sites
x=np.linspace(-2,2,19)
y=quickpiece(x)
plt.plot(x,y,'bo')

#Chunking and Averaging the data sites
n=2
xc=avg(chunks(x,n))
yc=avg(chunks(y,n))

#Defining our interpolation function from the data sites
rbf=Rbf(x,y, epsilon=0.2, function=my_function)
#rbfc=Rbf(xc,yc, epsilon=0.5)

#Applying the interpolation
xi=np.linspace(-2,2,100)
yi=rbf(xi)
#yic=rbfc(xi)
plt.plot(xi,yi,'g')
#plt.plot(xi,yic,'m')

#Nearest Neighbor Tests
datasites=zip(x,y)
tree=KDTree(datasites, leafsize=4)
points=np.array([[1.5,0]])
KNN=tree.query(points, k=3)[1][0]
#print datasites[KNN[1]]
for idx in KNN:
    plt.plot(datasites[idx][0],datasites[idx][1],'ro')
#The 'true' piecewise function
#plt.plot(xi,quickpiece(xi),'r')
```
