#Implementing Radial Basis Functions using SciPy

```python
#Importing and aliasing numpy and pyplot
import numpy as np
import matplotlib.pyplot as plt

#The SciPy interpolate library has a Radial Basis Function called Rbf
from scipy.interpolate import Rbf

#This allows us to display iPython Notebook plotting results inside the notebook
%pylab inline

#Named tuples for readability
from collections import namedtuple
```

###Radial Basis Interpolation 2D

```python
#Function to quickly generate a sampl piecewise function
def quickpiece(x):
    return np.piecewise(x,[x<1,x>=1],[0.5,1])
```

```python
#Named Tuple Creator
Points=namedtuple('Points', 'x y')

#Defining the data sites
x=np.linspace(0,2,19)
ds=Points(x,quickpiece(x))
plt.plot(ds.x,ds.y,'bo')

#Defining our interpolation function from the data sites
rbf=Rbf(ds.x,ds.y, epsilon=0.5)

#Applying the interpolation
xi=np.linspace(0,2,100)
interp=Points(xi, rbf(xi))
plt.plot(interp.x,interp.y,'g')

#The 'true' piecewise function
plt.plot(xi,quickpiece(xi),'r')
```

###Radial Basis Interpolation 3D

```python
from mpl_toolkits.mplot3d import Axes3D

def z_piece(X,Y):
    return np.maximum(quickpiece(X),quickpiece(Y))

fig = plt.figure()
ax = fig.add_subplot(111,projection='3d')

xi=np.linspace(0,2,200)
yi=np.linspace(0,2,200)
X,Y=np.meshgrid(xi,yi)


x=np.random.choice(xi,50)
y=np.random.choice(yi,50)
z=z_piece(x,y)
ax.plot(x,y,z, 'bo')






Z=z_piece(X,Y)

rbf=Rbf(x,y,z)
zi=rbf(X,Y)
ax.plot_surface(X,Y,zi, color='g')


ax.plot_surface(X,Y,Z,color='r',alpha=0.2)
```
