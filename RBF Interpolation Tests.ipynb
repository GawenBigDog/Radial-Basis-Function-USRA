```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import Rbf
%pylab inline
```

```python
x = np.linspace(0,10,9)
y = np.sin(x)
xi = np.linspace(0,10,100)
```

```python
rbf= Rbf(x,y) #Not sure why I need to include y as an argument here
yi=rbf(xi)
```

```python
plt.plot(xi,yi,'r')
plt.plot(xi,np.sin(xi),'g')
plt.plot(x, y, 'bo')
```

```python
def quickpiece(x):
    return np.piecewise(x,[x<1,x>=1],[0.5,1])
```

```python
x=np.linspace(0,2,9)
xi=np.linspace(0,2,100)
y=quickpiece(x)
#y=np.piecewise(x,[x<1,x>=1],[0.5,1])
plt.plot(x,y,'bo')
rbf=Rbf(x,y, epsilon=0.2)
yi=rbf(xi)
plt.plot(xi,yi,'g')
plt.plot(xi,quickpiece(xi),'r')
```
